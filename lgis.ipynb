```python
def longplus(seq):
    n = len(seq)
    dp = [1] * n
    prev = [-1] * n
    for i in range(1, n):
        for j in range(i):
            if seq[j] < seq[i] and dp[j] + 1 > dp[i]:
                dp[i] = dp[j] + 1
                prev[i] = j
    max_length = max(dp)
    idx = dp.index(max_length)
    lis = []
    while idx != -1:
        lis.append(seq[idx])
        idx = prev[idx]
    return lis[::-1]

def longmoins(seq):
    n = len(seq)
    dp = [1] * n
    prev = [-1] * n
    for i in range(1, n):
        for j in range(i):
            if seq[j] > seq[i] and dp[j] + 1 > dp[i]:
                dp[i] = dp[j] + 1
                prev[i] = j
    max_length = max(dp)
    idx = dp.index(max_length)
    lds = []
    while idx != -1:
        lds.append(seq[idx])
        idx = prev[idx]
    return lds[::-1]

with open("lgis.txt") as file:
    lines = file.readlines()
n = int(lines[0].strip())
pi = list(map(int, lines[1].strip().split()))

lis = longplus(pi)
lds = longmoins(pi)

print(" ".join(map(str, lis)))
print(" ".join(map(str, lds)))
```
